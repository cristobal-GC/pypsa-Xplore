# Rule: **retrieve_mobility_profiles**

**Description**

Data from BASt (Bundesanstalt für Straßenwesen, Instituto Federal Alemán de Investigación de Carreteras), of gauging stations on highways and federal roads from the years 2010-2014.

**Outputs**

- data/mobility_profiles/archive/2025-08-27/`kfz.csv`
- data/mobility_profiles/archive/2025-08-27/`pkw.csv`

```python
##### Import packages
import matplotlib.pyplot as plt
import os 
import sys


##### Import local functions
sys.path.append(os.path.abspath(os.path.join('..')))
import functions as xp


##### Read params.yaml
params = xp.read_params('../params.yaml')


##### Ignore warnings
import warnings
warnings.filterwarnings('ignore', category=UserWarning)
```

## `kfz.csv`

Contains weekly traffic profile for all motor vehicles (="Kraftfahrzeuge", i.e. cars, trucks, buses, motorcycles).

Load the file and preview its content.

```python
file = f"kfz.csv"

df_kfz = xp.load_file_csv(
    params,
    file,
    prefix='',
    name='',
    location="data/mobility_profiles/archive/2025-08-27/",
    skip_initial_lines=2,    
)

df_kfz.head()
```

```python
# Figure: count vs hour, one line per day (0-6)

df = df_kfz.copy()

pivot_count = (
    df.pivot_table(index="hour", columns="day", values="count", aggfunc="mean")
    .reindex(index=range(24), columns=range(7))
)

ax = pivot_count.plot(figsize=(10, 5), linewidth=2)
ax.set_xlabel("hour")
ax.set_ylabel("count")
ax.set_title("count by hour and day")
ax.legend(title="day", ncol=2, fontsize=8)
ax.set_xticks(range(24))
plt.tight_layout()
plt.show()
```
