```python
##### Import packages
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt


##### Import local functions
sys.path.append(os.path.abspath(os.path.join('..')))
import functions as xp


##### Read params.yaml
params = xp.read_params('../params.yaml')


##### Ignore warnings
import warnings
warnings.filterwarnings('ignore', category=UserWarning)
```

## `data/heat_load_profile_BDEW.csv`

Load the file from the PyPSA-Spain root path and show its structure.

```python
df = xp.load_file_csv(
    params,
    filename='heat_load_profile_BDEW.csv',
    location='data',
    index_col=0,
)

df
```

```python
groups = {
    'residential space': ['residential space weekday', 'residential space weekend'],
    'services space':    ['services space weekday',    'services space weekend'],
    'residential water': ['residential water weekday', 'residential water weekend'],
    'services water':    ['services water weekday',    'services water weekend'],
}

fig, axes = plt.subplots(2, 2, figsize=(14, 8), sharex=True)
for ax, (title, cols) in zip(axes.flat, groups.items()):
    for col in cols:
        label = 'weekday' if 'weekday' in col else 'weekend'
        ax.plot(df.index, df[col], label=label)
    ax.set_title(title)
    ax.set_xlabel(df.index.name or 'Hour')
    ax.set_ylabel('Load (p.u.)')
    ax.legend()

plt.suptitle('Heat load profiles BDEW', fontsize=13, y=1.01)
plt.tight_layout()
plt.show()
```
