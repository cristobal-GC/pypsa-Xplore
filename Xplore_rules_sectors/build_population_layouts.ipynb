# Rule: **build_population_layouts**

**Description**

Build mapping between cutout grid cells and population (total, urban, rural).

Input population data are at NUTS3 level. Thus, weather cells within the same NUTS3 region have the same population. 

Weather cells are classified between urban/rural according to its population density based on *urban_fraction* from World Bank. But the aforementioned limitation could lead to region-based classification, rather than weather_cell-based classification.

**Outputs**

- resources/`pop_layout_total.nc`
- resources/`pop_layout_urban.nc`
- resources/`pop_layout_rural.nc`

```python
######################################## Parameters

### Run
name = 'case_SectorCoupled'
prefix = ''

### Spatial domain 'ES' or 'EU' (for maps domain and NUTS regions)
spatial_domain = 'ES'

### Color for cells with exactly zero value in the first map
zero_value_color = "#9CC6D5"
```

```python
##### Import packages
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import cartopy.crs as ccrs
import os 
import sys
import numpy as np
import xarray as xr


##### Import local functions
sys.path.append(os.path.abspath(os.path.join('..')))
import functions as xp


##### Read params.yaml
params = xp.read_params('../params.yaml')


##### Ignore warnings
import warnings
warnings.filterwarnings('ignore', category=UserWarning)
```

## `pop_layout_total.nc`

Load the file and preview its content.

```python
file = f'pop_layout_total.nc'
path = f'{params["rootpath"]}/resources/{prefix}/{name}/'

ds = xr.open_dataset(path + file)

ds
```

```python
# Plot the single variable as a function of y (latitude) and x (longitude)
var_name = list(ds.data_vars)[0]
da = ds[var_name].squeeze()

domain = spatial_domain.upper()
bounds_key = f"boundaries_onshore_{domain}"
cmap_white_red = mcolors.LinearSegmentedColormap.from_list("white_red", ["white", "red"] )

if bounds_key not in params:
    raise KeyError(
        f"'{bounds_key}' was not found in params.yaml. "
        "Expected domains such as 'ES' or 'EU'."
    )

xmin, xmax, ymin, ymax = params[bounds_key]
total_mill = da.sum(skipna=True).item() / 1000

if {"y", "x"}.issubset(da.dims):
    fig, ax = plt.subplots(
        figsize=(8, 6),
        subplot_kw={"projection": ccrs.PlateCarree()},
    )
    da.plot(
        x="x",
        y="y",
        ax=ax,
        transform=ccrs.PlateCarree(),
        cmap=cmap_white_red,
        cbar_kwargs={"label": "thousands"},
    )

    # Overlay exactly-zero values with a user-selected color while preserving the base scale.
    zero_mask = xr.where(da == 0, 1.0, np.nan)
    zero_mask.plot(
        x="x",
        y="y",
        ax=ax,
        transform=ccrs.PlateCarree(),
        cmap=mcolors.ListedColormap([zero_value_color]),
        add_colorbar=False,
    )

    # Add map features
    xp.map_add_features(ax, params["map_add_features"])

    
    ax.set_title(
        f"Total population layout ({domain}) | total: {total_mill:.2f} Mill."
    )
    ax.set_xlabel("Longitude (x)")
    ax.set_ylabel("Latitude (y)")
    ax.set_extent([xmin, xmax, ymin, ymax], crs=ccrs.PlateCarree())
    plt.tight_layout()
    plt.show()
else:
    raise ValueError(
        f"Variable '{var_name}' does not have x/y dimensions. Found dims: {da.dims}"
    )
```
